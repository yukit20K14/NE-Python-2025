### List Operations and Methods

In Python, a list is a built-in dynamic sized array (automatically grows and shrinks). We can store all types of items (including another list) in a list.

- List can contain duplicate items.
- List in Python are Mutable. Hence, we can modify, replace or delete the items.
- List are ordered. It maintain the order of elements based on how they are added.
- Accessing items in List can be done directly using their position (index), starting from 0.

```python
# This is a list of fruits, a list has items that can be of any data type
# Each item is separated by a comma, and the list is encapsulated by square brackets
fruits = ['apple', 'banana', 'cherry']

# Python has a set of built-in methods that you can use on lists.
# Unpack by assigning variables to each item
a, b, c = fruits

# Accessing list item 
print(fruits[0]) # "apple"

# Change list item 
fruits[1] = "orange" # ['apple', 'orange', 'cherry']

# Add item to list
fruits.append("coconut") # ['apple', 'orange', 'cherry', 'coconut']
fruits.insert(1,"peach") # ['apple', 'peach', 'orange', 'cherry', 'coconut']

# Remove item from list
fruits.pop(3) # ['apple', 'peach', 'orange', 'coconut']
fruits.remove("apple") # ['peach', 'orange', 'coconut']

print(fruits)
```

```python
# Looping thruogh a list
for f in fruits:
    print(f)
```

```python
# If declaring a variable this way, any changes to the new variable is reflected on the initial one
# the new variable will only be a reference to the initial list
b = fruits
print(fruits)
b[0] = ""
print(fruits)
```

```python
# Deep copy, ensures that this does not occur
c = fruits.copy()
d = list(fruits)
e = fruits[:]
```

```python
# Join lists
fruits2 = ["coconut", "banana"]
fruits.extend(fruits2)
print(fruits)
```

```python
# Sorting lists
nums = [5, 7, 6, 0, 1, 4]
nums.sort()
print(nums)
# Reverse lists
nums.reverse()
print(nums)
```

### List Comprehensions

List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.


```python
# List Comprehension
# An example of how to create a list based on values in an existing list
fruits=[('apple','red'),('banana','yellow'),('cherry','red'),('mango','yellow')]
fruit_list=[]
for fruit,color in fruits:
    fruit_list.append(fruit) 
print(fruit_list)

# Done using list comprehensions
print([f for f, c in fruits]) # saves the need to write a for loop 

# Could also do flow control
print([fruit for fruit,color in fruits if color=='red'])

# What does this do? How does it work?
print([item for fruit in fruits for item in fruit][::2])
```

### Set characteristics and operations

- Set is a collection of distinct objects which is unordered, unchangeable, and unindexed.
- Set is written within curly brackets {}. 
- Set cannot have two items with the same value.



### Application of Sets in the real world 



1. E: Everyone
2. A: People love cat
3. B: People love dog

Set Operation
- `A | B` : People love cats or dogs
- `A - B` : People love cats only
- `B - A` : People love dogs only
- `A & B` : Intersection of people who loves both animals
- `(A | B) - (A & B)` : Mutual Exclusives

```python
A = {1, 2, 3, 4, 5}
B = {3, 5, 7}
print(A | B) # {1, 2, 3, 4, 5, 7}
print(A - B) # {1, 2, 4}
print(B - A) # {7}
print(A & B) # {3, 5}
print((A | B) - (A & B)) # {1, 2, 4, 7}
```

## Dictionary

### Objectives:

### Dictionary Characteristics and Operations

- Dictionary is a collection which is ordered, changeable and do not allow duplicates.
- Dictionary is written within curly brackets {key1:value1,key2:value2}. 
- Dictionary cannot have two items with the same key.



```python
# declaring a dictionary 
routers = {
    'PE1' : '10.0.0.1', 
    'PE2' : '10.0.0.2', 
    'PE3' : '10.0.0.3'
}

routers['PE1'] # finding values based on a key, returns '10.0.0.1'

routers['PE4'] = '10.0.0.4' # add new item

routers.keys() # retrieve list of keys
routers.values() # retrieve list of values
routers.items() # retrieve list of (key, value) pairs

# Iterating through values in a dictionary (no key)
for ip in routers.values(): 
	print(ip)
```

```python
# can be iterated using .items() or .keys() as well
for pe, ip in routers.items():
    print(f"{pe} - {ip}")
```

```python
routers.update({"PE5": "10.10.0.1"}) # update method add key, value for PE5
routers.pop("PE3")
print(routers.items())
```

#### JSON

JSON (JavaScript Object Notation)
- Data in key/value pairs
- String in "double quote" 
- Data is separate by commas
- Curly braces hold objects
- Square brackets hold arrays



```python
# when working with JSON, it is recommended to use the json module
import json

# given a JSON string 
json_string = '{"name":[{"name1":"value1"},{"name2":"value2"},{"name3":"value3"}]}'

json_obj = json.loads(json_string) # converts the JSON string to a dictionary
print(json_obj, type(json_obj))
json_obj["val"] = 100 # can perform dictionary operations after conversion

json_string_2 = json.dumps(json_obj) # converts the dictionary to a JSON string
print(json_string_2) # notice the added key and value added to the JSON string
```

#### POST and PATCH Requests

Use POST to create new record and PATCH to update existing record.

> `patch(url, data, args)`: Sends a PATCH request to the specified url
>
> `post(url, data, json, args)`: Sends a POST request to the specified url

```python
# Use POST to create new record and PATCH to update existing record.
import requests
dcim="https://demo.netbox.dev/api/dcim"

# Replace the API token with your own.
token={"Authorization": "Token c05338e234bdfcd9725e5c890759a5fa149fa9c3"}

# Find out the manufacturer id of Nokia.
payload={"manufacturer": 15, "model": "7750 SR-1", "slug": "sr1"} 
response=requests.post(f"{dcim}/device-types/", headers=token, json=payload)
device_id=response.json()['id'] #new device id created

# Update the comment field use patch method.
payload={"comments": "Test"} 
response=requests.patch(f"{dcim}/device-types/{device_id}/", headers=token, json=payload)
print(response.status_code)
```

### open() and os module

The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file:

> "r" - Read - Default value. Opens a file for reading, error if the file does not exist
>
> "a" - Append - Opens a file for appending, creates the file if it does not exist
>
> "w" - Write - Opens a file for writing, creates the file if it does not exist
>
> "x" - Create - Creates the specified file, returns an error if the file exists

```python
# OS Module is used for file handling.
import os # import built-in module os 
os.getcwd() # list current directory
os.listdir() # list all files
os.chdir('target') # change to target directory

# open file in read-only mode
with open("test.txt","r") as file: 
    data = file.read()
    print(data)

# open file in append mode, "w" overwrite mode
with open("test.txt","a") as file: 
    file.write(" Add some text.")
```
